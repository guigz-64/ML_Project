```python
import numpy as np
X = np.random.randint(2, size=(40,10))
Y = [sum(row) for row in X]
print(Y)
X_train = X[:30]
X_test = X[30:]
Y_train = Y[:30]
Y_test = Y[30:]
print("X_train",X_train)
print("X_test",X_test)
print("Y_train",Y_train)
print("Y_test",Y_test)
```

```python
import matplotlib.pyplot as plt
import random
import math
```

```python
#1
class Neural_Network():
    def init(self,x,y):
        #petite valeur pour éviter d'avoir des valeurs très grandes de Vf et Vx au fil des itérations(et donc un programme qui crash)
        self.Vx=x
        self.Vf=y
       
        
    def rec_relation(self,xk,f):
        #fonction retourne la valeur f+1
        return xk*self.Vx + f*self.Vf
    
    def forward_propa(self,x):
        #f0 vaut 0
        f = np.zeros((x.shape[0],x.shape[1]))
        
        
        for k in range(0,x.shape[1]-1):
            #calcul de l'ensemble des valeurs f en fonction de la 30x10 de X
            #retourne une matrice 30x10
            f[:,k+1]= NN.rec_relation(x[:,k],f[:,k])
        
        return f
    
    def back_propa(self,y,x,iteration,alpha):
        for z in range(iteration):
            temp_Vx=0
            temp_Vf=0
            dEdVx=0
            dEdVf=0
            y_barre=['']*30
            
            #appel de la fonction de récurrence 
            ft=NN.forward_propa(x)
            
            for m in range(0,x.shape[0]):
                #obtention de y chapeau en ajoutant uniquement la dernière colonne de la matrice ft car y_barre = fT avec T=10
                y_barre[m]=ft[m,9]
            
        
            for l in range(0,10):
                #calcul de l'erreur
                e=1 / 2 * np.sum((y_barre[l] - y[l])**2)
            
            
                
            
               
            #calcul des dérivé partiel de E en fonction de Vx et Vf        
            for t in range (1,x.shape[1]):
                for i in range (1,x.shape[0]):
                    temp_Vx=temp_Vx+(y_barre[i]-y[i])*x[i,t]
                    temp_Vf=temp_Vf+(y_barre[i]-y[i])*ft[i,t-1]
                dEdVx=dEdVx+temp_Vx*self.Vf**(10-t)
                dEdVf=dEdVf+temp_Vf*self.Vf**(10-t)
            
            #on améliore a chaque itération les valeurs de Vx et Vf pour obtenir un y chapeau qui colle le plus aux valeurs réels
            self.Vx=self.Vx-alpha*dEdVx
            self.Vf=self.Vf-alpha*dEdVf
            
            if(z==iteration-1):
                print("meilleur erreur",e)
            #on plot l'erreur qui permet de comparer les valeurs de notre modèle aux valeurs réels 
            plt.scatter(z,e)
            
            
    #on utilise la même base que la fonction back_propa
    def resilient_propa(self,y,x,iteration):
        temp_x=0
        temp_f=0
        delta_x=0.001
        delta_f=0.001
        for z in range(iteration):
            temp_Vx=0
            temp_Vf=0
            dEdVx=0
            dEdVf=0
            np1=1.2
            nn=0.5
            y_barre=['']*30
            
                
            ft=NN.forward_propa(x)
            
            for m in range(0,x.shape[0]):
                y_barre[m]=ft[m,9]
            
            for l in range(0,10):
                e=1 / 2 * np.sum((y_barre[l] - y[l])**2)
                
            for t in range (1,x.shape[1]):
                for i in range (1,x.shape[0]):
                    temp_Vx=temp_Vx+(y_barre[i]-y[i])*x[i,t]
                    temp_Vf=temp_Vf+(y_barre[i]-y[i])*ft[i,t-1]
                dEdVx=dEdVx+temp_Vx*self.Vf**(10-t)
                dEdVf=dEdVf+temp_Vf*self.Vf**(10-t)
            
            #compaaraison par rapport au signe de la valeur de dEdVx actuelle et précèdente 
            if(np.sign(temp_x)==np.sign(dEdVx)):
                delta_x=delta_x*np1
            else:
                delta_x=delta_x*nn
            
            if(np.sign(temp_f)==np.sign(dEdVf)):
                delta_f=delta_f*np1
            else:
                delta_f=delta_f*nn
                
            #on améliore cette fois Vx et Vf d'une autre manière que pour l'algorithme de back_propa
            #ça évite les problèmes de vanishing et explosion problem
            self.Vx=self.Vx-np.sign(dEdVx)*delta_x
            self.Vf=self.Vf-np.sign(dEdVf)*delta_f
             
            #on actualise les nouvelles valeurs des dérivées 
            temp_x=dEdVx
            temp_y=dEdVf
            
            if(z==iteration-1):
                print(" meilleur erreur :",e)
                
            plt.scatter(z,e)
        
            
    def norme(self,x,y):
        #calcul norme vectorielle
        return math.sqrt((x**2)+(y**2))
    
    #base similaire aux deux autres algorithmes
    def gradient_clipping(self,y,x,iteration,alpha,n):
        for z in range(iteration):
            temp_Vx=0
            temp_Vf=0
            dEdVx=0
            dEdVf=0
            y_barre=['']*30
            
           
            ft=NN.forward_propa(x)
            
            for m in range(0,x.shape[0]):
                y_barre[m]=ft[m,9]
            
        
            for l in range(0,10):
                e=1 / 2 * np.sum((y_barre[l] - y[l])**2)
            #print("erreur",e)   
                    
            for t in range (1,x.shape[1]):
                for i in range (1,x.shape[0]):
                    temp_Vx=temp_Vx+(y_barre[i]-y[i])*x[i,t]
                    temp_Vf=temp_Vf+(y_barre[i]-y[i])*ft[i,t-1]
                dEdVx=dEdVx+temp_Vx*self.Vf**(10-t)
                dEdVf=dEdVf+temp_Vf*self.Vf**(10-t)
            
            #on calcul d'une nouvelles manière le gradient de E
            #la norme du gradient doit être suppérier à éta
            if(self.norme(dEdVx,dEdVf)>n):
                self.Vx=self.Vx-alpha*(n/self.norme(dEdVx,dEdVf))*dEdVx
                self.Vf=self.Vf-alpha*(n/self.norme(dEdVx,dEdVf))*dEdVf
            
            if(z==iteration-1):
                print("meilleur erreur :",e)
            
            plt.scatter(z,e)
        
        
        
```

```python
NN = Neural_Network()
#dans l'ensemble des plots l'erreur est en ordonnée
# si Vf et Vx trop élevé : phénomène de vanishing and explosion problem 
NN.init(0.1,0.1)

#si  alpha trop élevé : phénomène de vanishing and explosion problem 
NN.back_propa(Y_train,X_train,500,0.0001)
```

```python
NN.init(0.1,0.1)
#à partir de la 43e itération la courbe de l'erreur diverge
#ce problème vient du fait que l'algorithme à trouver l'erreur la plus basse et les valeurs les plus proche 
#des valeurs de Y_train 
NN.resilient_propa(Y_train,X_train,43)
```

```python
NN.init(0.1,0.1)
NN.gradient_clipping(Y_train,X_train,150,0.1,0.1)
```
